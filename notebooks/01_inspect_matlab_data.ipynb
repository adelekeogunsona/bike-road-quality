# Inspect MATLAB Sensor Data

Quick look at the exported CSV data to understand what we're working with.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

DATA_DIR = '../data/sample'

pos = pd.read_csv(f'{DATA_DIR}/position.csv', parse_dates=['timestamp'])
acc = pd.read_csv(f'{DATA_DIR}/acceleration.csv', parse_dates=['timestamp'])
mag = pd.read_csv(f'{DATA_DIR}/magnetic_field.csv', parse_dates=['timestamp'])
ori = pd.read_csv(f'{DATA_DIR}/orientation.csv', parse_dates=['timestamp'])
gyro = pd.read_csv(f'{DATA_DIR}/angular_velocity.csv', parse_dates=['timestamp'])

print(f'Position: {len(pos)} samples')
print(f'Acceleration: {len(acc)} samples')
print(f'Magnetic field: {len(mag)} samples')
print(f'Orientation: {len(ori)} samples')
print(f'Angular velocity: {len(gyro)} samples')
```

```python
# Position data overview
print('Position columns:', list(pos.columns))
print()
pos.describe()
```

```python
# Plot the raw route
plt.figure(figsize=(8, 8))
plt.plot(pos['longitude'], pos['latitude'], 'b-', linewidth=0.5)
plt.plot(pos['longitude'].iloc[0], pos['latitude'].iloc[0], 'go', markersize=10, label='Start')
plt.plot(pos['longitude'].iloc[-1], pos['latitude'].iloc[-1], 'ro', markersize=10, label='End')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Raw GPS Route')
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Acceleration magnitude
acc['magnitude'] = np.sqrt(acc['ax']**2 + acc['ay']**2 + acc['az']**2)

fig, axes = plt.subplots(2, 1, figsize=(12, 6))

time_sec = (acc['timestamp'] - acc['timestamp'].iloc[0]).dt.total_seconds()
axes[0].plot(time_sec, acc['magnitude'], linewidth=0.3)
axes[0].set_ylabel('Acc magnitude (m/s²)')
axes[0].set_title('Acceleration over time')
axes[0].axhline(y=9.81, color='r', linestyle=':', alpha=0.5)

gyro_time = (gyro['timestamp'] - gyro['timestamp'].iloc[0]).dt.total_seconds()
gyro_mag = np.sqrt(gyro['wx']**2 + gyro['wy']**2 + gyro['wz']**2)
axes[1].plot(gyro_time, gyro_mag, linewidth=0.3, color='purple')
axes[1].set_ylabel('Gyro magnitude (rad/s)')
axes[1].set_xlabel('Time (seconds)')
axes[1].set_title('Angular velocity over time')

plt.tight_layout()
plt.show()
```

```python
# Check horizontal accuracy
plt.figure(figsize=(10, 3))
pos_time = (pos['timestamp'] - pos['timestamp'].iloc[0]).dt.total_seconds() / 60
plt.plot(pos_time, pos['horizontal_accuracy'])
plt.xlabel('Time (minutes)')
plt.ylabel('Horizontal accuracy (m)')
plt.title('GPS Horizontal Accuracy Over Time')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f'Mean accuracy: {pos["horizontal_accuracy"].mean():.1f} m')
print(f'Worst accuracy: {pos["horizontal_accuracy"].max():.1f} m')
```
